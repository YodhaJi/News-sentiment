```python
import requests
import pandas as pd
import numpy as np
import newsapi
from bs4 import BeautifulSoup as bs
from newspaper import Article
import nltk
import joblib
```

```python
api_key ="API-key"
from newsapi import NewsApiClient
newsapi = NewsApiClient(api_key=api_key)
```

```python
search = "[SEARCH NEWS]"
top_headlines = newsapi.get_everything(q=search, language='en')
```

```python
data = pd.DataFrame(top_headlines['articles'])[["title","author","publishedAt","url"]]
```

```python
data["publishedAt"]=pd.to_datetime(data["publishedAt"]).dt.tz_convert('Asia/Kolkata').dt.strftime('%e %B,%Y')
```

```python
def extracting_test(x):
  response = requests.get(x)
  soup = bs(response.text)
  return soup.get_text()
```

```python
data["text"]=data["url"].apply(extracting_test)
```

```python
data['text'].values
import joblib
vectorizer = joblib.load('/content/drive/MyDrive/Colab Notebooks/Projects/News/vectorizer.pkl')
```

```python
from tensorflow.keras.models import load_model
model = load_model('/content/drive/MyDrive/Colab Notebooks/Projects/News/themodel.h5')
```

```python
pred= model.predict(vectorizer.transform(data['text'].values).toarray()).round()
```

```python
result = pd.concat([data,pd.DataFrame(pred)],axis=1)
```
